# Passo 1: importar as bibliotecas necessárias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os as os
from google.colab import files
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

# Passo 2: Carregar o dataset

```python
# Carregar o dataset Titanic
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
# Resumo das colunas
print("\nResumo das colunas:")
print(df.info())
```

```python
# Exibir as primeiras linhas do dataset
print(df.head())
```

```python
# Verificar valores ausentes
print(df.isnull().sum())
```

```python
# Salvar o DataFrame como um arquivo CSV
df.to_csv('titanic_dados_originais.csv', index=False)
```

```python
# Baixar o arquivo CSV
files.download('titanic_dados_originais.csv')
```

# Passo 3: Tratar Valores Ausentes

```python
# Preencher valores ausentes em colunas numéricas com a mediana
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
# Verificar valores ausentes
print(df.isnull().sum())
```

```python
# Preencher valores ausentes em colunas categóricas com a moda
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
# Verificar valores ausentes
print(df.isnull().sum())
```

```python
# Salvar o DataFrame como um arquivo CSV
df.to_csv('titanic_dados_valores_ausentes.csv', index=False)
```

```python
# Baixar o arquivo CSV
files.download('titanic_dados_valores_ausentes.csv')
```

# Passo 4: Codificar Variáveis Categóricas

```python
# Codificar variáveis categóricas
encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_cols = pd.DataFrame(encoder.fit_transform(df[['Sex', 'Embarked']]),
                            columns=encoder.get_feature_names_out(['Sex', 'Embarked']))
df = pd.concat([df, encoded_cols], axis=1)
df.drop(['Sex', 'Embarked'], axis=1, inplace=True)
```

```python
# Exibir as primeiras linhas do dataset
print(df.head())
```

- Coluna Sex_male: Indica se o passageiro é do sexo masculino (1) ou feminino (0).
- Coluna Embarked_Q: Indica se o passageiro embarcou em Queenstown (1) ou não (0).
- Coluna Embarked_S: Indica se o passageiro embarcou em Southampton (1) ou não (0).

```python
# Salvar o DataFrame como um arquivo CSV
df.to_csv('titanic_dados_variaveis_categoricas.csv', index=False)
```

```python
# Baixar o arquivo CSV
files.download('titanic_dados_variaveis_categoricas.csv')
```

# Passo 5: Normalização das colunas numéicas

```python
# Selecionar colunas numéricas
numerical_features = ['Age', 'Fare', 'SibSp', 'Parch']
```

```python
# Mostrar as estatísticas antes da normalização
print("Estatísticas antes da normalização:")
print(df[numerical_features].describe())
```

O efeito visual da normalização usando StandardScaler é transformar os valores das colunas numéricas selecionadas para que tenham média 0 e desvio padrão 1. Isso torna os dados mais comparáveis e compatíveis para análises estatísticas e modelos de aprendizado de máquina, ajudando a melhorar o desempenho e a precisão dos modelos.

```python
# Normalizar dados numéricos
scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])
```

```python
# Mostrar as estatísticas após a normalização
print("\nEstatísticas após a normalização:")
print(df[numerical_features].describe())
```

```python
# Visualizar as primeiras linhas do DataFrame transformado
print("\nPrimeiras linhas do DataFrame transformado:")
print(df.head())
```

```python
# Salvar o DataFrame como um arquivo CSV
df.to_csv('titanic_dados_normalizados.csv', index=False)
```

```python
# Baixar o arquivo CSV
files.download('titanic_dados_normalizados.csv')
```
